There are three distinct numeric types: integers, floating point numbers, and complex numbers. In addition, Booleans are a subtype of integers. Integers have unlimited precision.

```python
x = 2
y = 3
print("x + y:", x+y)
print("x - y:", x-y)
print("x * y:", x*y)
print("x / y:", x+y)
print("x//y:", x//y)
print("x%y:", x%y)
print("-x:", -x)
print("+x:", x%y)
print("absolute :", abs(3.4))
print("Cast to INT :", int(3.4))
print("Cast to FLOAT :", float(3))
print("Complex Number :", complex(2,3))
print("2 Rasied to 3 :", pow(2,3))
print("2 Rasied to 3 :", 2**3)
```

The result of bitwise operations is calculated as though carried out in two’s complement with an infinite number of sign bits.

The priorities of the binary bitwise operations are all lower than the numeric operations and higher than the comparisons; the unary operation ~ has the same priority as the other unary numeric operations (+ and -).

```python
x = 2
y = 3
print('x | y:',x | y)
print('x ^ y:',x ^ y)
print('x & y:',x & y)
print('x << n:',x << y)
print('x >> n',x >> y)
print('~x',~x)
```

## 2.Control Flow Statements

## 1. While Statement : Repeated until the statement is true
    1. Break Statement: exits the loop immediately
    2. Continue Statment: Skips the current iteration and proceeds to the next one

```python
a = 0
while a < 3:
    a += 1
    print("Value of a:", a)
    if a <= 1:
        continue
        a += 2
        print("Value of a:", a)
    else:
        print("Break Execution")
        break
        
```

#### List Comprehension:
* They provide a concise way of creating a list, like to square the each term in the list.
* A typcial syntax is: An expression which is followed by a for clause, then 0 or more for or if clause in a bracket and it results in a new list

```python
print("This is the list comprehension:", [x+y for x in [1,2] for y in [2,3]])
```

### Nested List Comprehension:
It provides a way to work with the multidimensional data in a single, compact expression.It combines for loops within a list comprehension and return a new list by applying an operation to each element.

```python
print("For example consider a matrix:")
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

[[row[i] for row in matrix] for i in range(3)]
```

### Merge, Join, Concat and Compare

```python
## Concatenates two or more dataframes by default axis is 0, indicates concat should happen vetically
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)


pd.concat([df1,df2,df3])

```
